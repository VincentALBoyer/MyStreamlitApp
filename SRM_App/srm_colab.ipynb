```python
# Step 2: Start Application
from google.colab import output
import subprocess
import time

# 1. Port settings
PORT = 8501

# 2. Start Streamlit in the background
print("üöÄ Starting Streamlit server...")
# Note: server.address localhost and headless true are critical for Colab
process = subprocess.Popen([
    "streamlit", "run", "app.py", 
    "--server.port", str(PORT), 
    "--server.address", "localhost", 
    "--server.headless", "true",
    "--browser.gatherUsageStats", "false"
])

# 3. Wait for server to initialize
time.sleep(5)

# 4. Serve the port as an iframe inside the notebook
print("‚ú® SRM Simulation Loading Below...")
output.serve_kernel_port_as_iframe(PORT, width='100%', height='800')
```
